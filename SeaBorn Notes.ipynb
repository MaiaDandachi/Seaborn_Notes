# Destirbution plots : displot , jointplot, pairplot, kdeplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Displot shows teh destibution of a part from the data 
```

```python
sns.distplot(tips['total_bill']) # we can add kde false to git rid of the line 
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
# JointPlot allows you to join 2 displots of the data with diff types of parameters : 
# “scatter” “reg” “resid” “kde” “hex”

sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
# Pairplot plots the relation  across the entire dataframe
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
# KDEplots :  KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value.
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

# Categorial plots

```python
# Bar plot :
#barplot is a general plot that allows you to aggregate the categorical data based off some function, by default the mean
```

```python
sns.barplot(x='time',y='total_bill',data=tips) # dinner avg totalbill is higher than lunch total bill
```

```python
# We can use another estimator function like standard deviantion to check the variance 
import numpy as np 
sns.barplot(x='time',y='total_bill',data=tips,estimator=np.std) # here the variance of the dinner total bill is higher
```

```python
# Count Plot counts the occurences 
sns.countplot(x='sex',data=tips) # it takes only x ases
```

```python
# Box Plot : divided into qurters and every point out the whiskers are outliers and in the 2 qurter there is the median 
# hue adds a categorial data 

sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow') # shows the destirbution of bills for each day
```

```python
# distribution of smokers and non smokers total bill in the days 
# Here on sunday the total bill of smokers is higher than than the non smokers where the avg of the total bill of smokers
# is 25 and non smokers almost 20
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips) 
```

```python
# ViolinPlot
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1')
```

```python
# stripplot and swarmplot
```

```python
# stripplot plots one categorial 
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
# jitter is used when we have alot of points overlapping so we can look at them easier
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True) 
```

```python
# hue : to categorize the data
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1')
# add split so it would be mor clear
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',split=True)
```

```python
# swarm plot doesnt not overlap the data points
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", split=True)
```

```python
# Combining Categorical Plots
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=5)
```

We should make our data in matrix form where our index value and column value matches

# Heatmap



```python
# we use correaltion to  make the data in matrix form so we can use heatmap
#the sns.heatmap function basically just colors it in for you.
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
tips = sns.load_dataset('tips')
```

```python
# Matrix form for correlation data
tips.corr()
```

```python
sns.heatmap(tips.corr(),cmap='coolwarm',annot=True)
```

```python
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
pvflights = flights.pivot_table(values='passengers',index='month',columns='year')
sns.heatmap(pvflights)
```

```python
sns.heatmap(pvflights,cmap='magma',linecolor='white',linewidths=1)
```

```python
# More options to get the information a little clearer like normalization
sns.clustermap(pvflights,cmap='coolwarm',standard_scale=1)
```

# Grid plot

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Pair grid

# Pairgrid is a subplot grid for plotting pairwise relationships in a dataset.

# Just the Grid
sns.PairGrid(iris)
```

```python
# Map to upper,lower, and diagonal
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(tips)
```

```python
# Just the Grid

g = sns.FacetGrid(tips, col="time", row="smoker")
g = g.map(plt.hist, "total_bill")
# g = g.map(plt.scatter, "total_bill", 'tip')
```

# Regresion plot

```python
# lmplot allows you to display linear models, but it also conveniently allows you to split up those
#plots based off of features, as well as coloring the hue based off of features.

sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100})
```

```python
# **** Using a Grid  ***
# We can add more variable separation through columns and rows with the use of a grid. 
#Just indicate this with the col or row arguments
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',aspect= 3 ,height = 2)
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex', palette = 'coolwarm')
```

```python
# we can add a row also , we can add palette for color = 'coolwarm'
# each column is either lunch or Dinner and the row is mae or female and displaying total bill and tip
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips, palette = 'coolwarm')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.6,size=8)
```
